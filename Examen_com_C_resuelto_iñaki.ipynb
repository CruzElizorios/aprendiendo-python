3. Escribir una función que reciba una temperatura en grados Fahrenheit y devuelva dicha temperatura expresada en grados Celcius. Si se recibe un valor menor al cero absoluto (- 459,67 °F), no debe realizar la conversión, ni retorar ningún valor (None).

  Recuerde que si F es una temperatura en grados Fahrenheit y C la misma temperatura en grados Celcius, existe la siguiente relación entre ambas:

      *C = (F − 32)/1, 8*

```python
def deFahrenCelc(fah: float) -> float:
  """
  Toma un numero en grados fahrenheit, realiza las cuentas para transformarlo en grados celcius y devuelve este último
  siempre que la misma sea igual o mayor al cero absoluto
  """
  if(fah < -459.67):
    return None

  gradosCel = (fah-32)/1.8
  return gradosCel
deFahrenCelc(-459.67)
```

4. Escribir una función que dada dos listas de igual dimensión, las recorra y para cada posición almacene el elemento de la primer lista como clave de un diccionario y el elemento de la segunda lista como valor.

```python
def crearDic(list1:list[str],list2:list[int])-> dict[str:int]:
  """
  Esta función crea un diccionario a partir de dos listas que deben tener la misma dimensión y este diccionario cuenta como clave a
  los valores de la primera lista y como valor los elementos de la segunda lista; si hay repetidos en la primer lista se queda con el
  último valor que aparezca
  """
  dic = {}
  if(len(list1) == len(list2)):

    for index in range(len(list1)):
      # for index2 in range(len(list2)):
      #   if(index == index2):
      #     dic[list1[index]] = list2[index2]
      dic[list1[index]] = list2[index]
    return dic
  else:
    return print(f"Las listas deben tener el mismo tamaño, estas son distintas: \ntamaño lista 1: {len(list1)} y tamaño lista 1: {len(list2)}")

print(crearDic(["A","B","A","C"], [1,2,3,4]))

print(crearDic(["A","B","A"], [1,2]))
```

```python
# código más simple (no tiene la verificación), el doble for no era necesario
def dos_listas_a_diccionario (claves: list, valores: list) -> dict:
  """
  Función que recibe dos listas de igual dimensión, las recorra y para cada
  posición almacena el elemento de la primer lista como clave de un diccionario
  y el elemento de la segunda lista como valor.
  Si hubiera valores repetidos en la primer lista, considera el último en orden
  de aparición.
  """
  diccionario = {}
  for posicion in range(len(claves)):
    diccionario[claves[posicion]] = valores[posicion]
  return diccionario
```

5. Escribir un programa que permita el ingreso de temperaturas registradas en grados Fahrenheit en diferentes barrios de la ciudad. Un ejemplo de posibles datos son los de la siguiente planilla:

  image.png

  Dado que la información es registrada a medida que se obtiene la medición y que no se conoce cuantas mediciones se realizan, debe proponerse un fin de datos adecuado.

  Una vez que finaliza el ingreso de datos, el programa debe mostrar en pantalla la última temperatura registrada en cada barrio de la ciudad, expresada en grados Celcius.


```python
def nuevoBarrio() -> tuple[str,list[int]]:
  """
  Le pide al usuario el nombre de un barrio y luego las temperaturas que desea agregar en °F (verifica que no haya ingresado letras,
  en tal caso vuelve a pedir la temp), esto retorna una tupla con los valores del nombre del barrio y su temperatura
  """
  listTemp = []
  letras = ["a","b","c","d","e","f","g","h","i","j","k","l","m","n","o","p","q","r","s","t","u","v","w","x","y","z"]
  i=1
  barrio = input("Ingrese el barrio: ")
  while True:
    temp = input(f"ingrese la {i}° temperatura en °F para el barrio {barrio} o x para terminar: ")
    if(temp == "x" or temp == "X"):
      break
    # desarma la palabra o el numero en cada componente
    for letra in temp.lower():
      #si hay una letra,el in devuelve True, por lo que frena el ciclo for y vuelve a pedir una temp
      if(letra in letras):
        break
    #despues de no encontrar letras; entra, convierte a un entero el dato ingresado en temp y lo agrega a la lista
    else:
      listTemp.append(int(temp))
      i+=1

  return (barrio,listTemp)

def cargarBarrios() -> list[tuple[str,list]]:
  """
  Le pregunta al usuario si quiere agregar un barrio y luego llama a la función nuevoBarrio, donde el usuario completa
  con el nombre del barrio y las temperaturas; estás se almacenan en una lista y es lo que devuelve esta función
  """
  barrios = []

  while True:
    control = input("Desea agregar un barrio nuevo? si / no : ")
    if(control in ["no","No","NO","N","n"]):
      break
    barrio = nuevoBarrio()
    barrios.append(barrio)
  return barrios

def principal() -> None:
  """
  Llama a la función cargarBarrios para pedirle los datos al usuario, con estos convierte en grado celcius la última
  temperatura registrada y la muestra por pantalla
  """
  listaBarrios = cargarBarrios()

  for barrio, temps in listaBarrios:
    ultTemp = temps[-1]
    graCel = deFahrenCelc(ultTemp)
    print(f"Para el barrio {barrio}, temperatura: {graCel}°C")
  return

principal()
```
